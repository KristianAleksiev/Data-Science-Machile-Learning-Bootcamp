## Divergence and overflow
## $g(x) = x^5 - 2x^4 + 2 $

```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
# Make data
x = np.linspace(start=-2.5, stop=2.5, num=1000)
```

```python
def f(x):
    return x ** 5 - 2 * x ** 4 + 2
def deriv_f(x):
    return 5 * x ** 4 - 8 * x ** 3
```

```python
def gradient_descent(derivative_func, initial_guess, multiplier=0.02, precision=0.001, max_iter=500):
    new_x = initial_guess
    x_values = [new_x]
    slope_values = [derivative_func(new_x)]

    for n in range(max_iter):
        previous_x = new_x
        gradient = derivative_func(previous_x)
        new_x = previous_x - multiplier * gradient

        step_size = abs(new_x - previous_x)

        x_values.append(new_x)
        slope_values.append(derivative_func(new_x))

        if step_size < precision:
            break

    return new_x, x_values, slope_values
```

```python
# Plotting function and derivative and scatter plot
local_min, list_x, deriv_list = gradient_descent(derivative_func=deriv_f, initial_guess=-0.2, max_iter=70)
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.title("Cost function", fontsize=17)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)
plt.plot(x, f(x), color="blue", linewidth=3, alpha=0.6)


plt.scatter(list_x, f(np.array(list_x)), color="red", s=100, alpha=0.6)

plt.subplot(1, 2, 2)
plt.title("Slope of the cost function", fontsize=17)
plt.xlabel("X", fontsize=16)
plt.ylabel("g' (x)'", fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)

plt.plot(x, deriv_f(x), color="skyblue", linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.6)            
plt.show()

print(f"Local min occurs at: {local_min}")
print(f"Cost at this minimum is: {f(local_min)}")
print(f"Number of steps: {len(list_x)}")
```

```python
# Plotting function and derivative and scatter plot
n=300
local_min, list_x, deriv_list = gradient_descent(derivative_func=deriv_f, initial_guess=0.5, multiplier=0.0025,
                                                 max_iter=n,
                                                 precision=0.001)
plt.figure(figsize=[15, 5])

plt.subplot(1, 2, 1)

plt.xlim(-1.2, 2.5)
plt.ylim(-1, 4)
plt.title("Cost function", fontsize=17)
plt.xlabel("X", fontsize=16)
plt.ylabel("f(x)", fontsize=16)
plt.plot(x, f(x), color="blue", linewidth=3, alpha=0.6)


plt.scatter(list_x, f(np.array(list_x)), color="red", s=100, alpha=0.6)

plt.subplot(1, 2, 2)
plt.title("Slope of the cost function", fontsize=17)
plt.xlabel("X", fontsize=16)
plt.ylabel("g' (x)'", fontsize=16)
plt.grid()
plt.xlim(-1, 2)
plt.ylim(-4, 5)

plt.plot(x, deriv_f(x), color="skyblue", linewidth=5, alpha=0.6)
plt.scatter(list_x, deriv_list, color="red", s=100, alpha=0.6)            
plt.show()

print(f"Local min occurs at: {local_min}")
print(f"Cost at this minimum is: {f(local_min)}")
print(f"Number of steps: {len(list_x)}")
```
