## Notebook imports and packages

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

## Example 1 - A simple cost function
## $f(x) = x^2 + x + 1$ - LaTeX mathematical notation

```python
def f(x):
    return x ** 2 + x + 1
```

```python
# Making the data
x_1 = np.linspace(start=-3, stop=3, num=100)
x_1
```

```python
plt.xlim(-3, 3)
plt.ylim(0,8)
plt.xlabel("x", fontsize=15)
plt.ylabel("f (x)", fontsize=15)
plt.plot(x_1, f(x_1))
plt.show()
# Lowest cost is at the lowest point of the graph x=-0.5, f(x)=1
```

## Slope and derivatives <=(f (x))

## $x^n = n * x ^ n-1$

$x^3 + 5x^2 + 7 = 3x^2 + 5(2x) = 3x^2 + 10x$ 

```python
def deriv_f(x):
    return 2*x + 1
```

## Plot function and derivative side by side:


```python
plt.figure(figsize=[15, 5])
# Chart one: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(0,8)
plt.title("Cost function", fontsize=17)
plt.xlabel("x", fontsize=15)
plt.ylabel("f (x)", fontsize=15)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)


# Chart two: Derivative function
plt.subplot(1, 2, 2)
plt.title("Sloap of cost function - Derivative", fontsize=15)
plt.xlabel("x", fontsize=15)
plt.ylabel("f' (x)", fontsize=15)

plt.grid()

plt.xlim(-2 ,3)
plt.ylim(-3 ,6)
plt.plot(x_1, deriv_f(x_1), color="skyblue", linewidth=5)
plt.show()
```

## Changing the charts to overlap

```python
plt.figure(figsize=(10, 10))
# Chart one: Cost function
plt.subplot(2, 1, 1)

plt.xlim(-3, 3)
plt.ylim(0,8)
plt.title("Cost function", fontsize=17)
plt.xlabel("x", fontsize=15)
plt.ylabel("f (x)", fontsize=15)
plt.plot(x_1, f(x_1), color="blue", linewidth=3)


# Chart two: Derivative function
plt.subplot(2, 1, 1)
plt.title("Intersection", fontsize=15)
plt.xlabel("x", fontsize=15)
plt.ylabel("f' (x)", fontsize=15)

plt.grid()

plt.xlim(-2 ,3)
plt.ylim(-3 ,6)
plt.plot(x_1, deriv_f(x_1), color="skyblue", linewidth=5)
plt.show()
```

# Gradient Descent

```python
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001
x_values = [new_x]
slope_list = [deriv_f(new_x)]
for n in range(50):
    previous_x = new_x
    gradient = deriv_f(previous_x)
    new_x = previous_x - step_multiplier * gradient
    
    step_size = abs(new_x - previous_x)
    
    x_values.append(new_x)
    slope_list.append(deriv_f(new_x))
    
    if step_size < precision:
        print(f"Loop ran {n} times")
        break
    
print(f"Local minimum occurs at: {new_x:.3f}")
print(f"Slope or deriv of f(x) value at this point is: {deriv_f(new_x)}")
print(f"f(x) value or cost at this point is: {f(new_x):.3f} \n")
print("X values : ", *x_values,"\n")
print("Slope values: ", *slope_list)
```

```python
# Plot Gradient descent calculations

plt.figure(figsize=(15, 5))
# Chart one: Cost function
plt.subplot(1, 2, 1)

plt.xlim(-3, 3)
plt.ylim(0,8)
plt.title("Cost function", fontsize=17)
plt.xlabel("x", fontsize=15)
plt.ylabel("f (x)", fontsize=15)
plt.plot(x_1, f(x_1), color="blue", linewidth=3, alpha=0.6)

values = np.array(x_values)
plt.scatter(x_values, f(values), color="red", s=100, alpha=0.6)

# Chart two: Derivative function
plt.subplot(1, 2, 2)
plt.title("Slope of the cost function", fontsize=15)
plt.xlabel("x", fontsize=15)
plt.ylabel("f' (x)", fontsize=15)

plt.grid()

plt.xlim(-2 ,3)
plt.ylim(-3 ,6)
plt.plot(x_1, deriv_f(x_1), color="skyblue", linewidth=5, alpha=0.6)
plt.scatter(x_values, slope_list, color="red", s=100, alpha=0.6)

plt.show()
```
